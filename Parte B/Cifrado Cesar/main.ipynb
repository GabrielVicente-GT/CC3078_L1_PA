```python
def decrypt_cesar(ciphertext, shift, alphabet):
    decrypted_text = ""
    for char in ciphertext:
        if char in alphabet:
            char_index = alphabet.index(char)
            shifted_char = alphabet[(char_index - shift) % len(alphabet)]
            decrypted_text += shifted_char
        else:
            decrypted_text += char
    return decrypted_text

def brute_force_decrypt(ciphertext, alphabet):
    for shift in range(len(alphabet)):
        decrypted_text = decrypt_cesar(ciphertext, shift, alphabet)
        print(f"Intento de descifrado con desplazamiento {shift}: {decrypted_text}")

alphabet = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"

with open("cipher1.txt", "r") as file:
    ciphertext = file.read()

brute_force_decrypt(ciphertext, alphabet)
```
