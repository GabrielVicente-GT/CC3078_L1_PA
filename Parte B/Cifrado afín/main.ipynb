```python
def mod_inverse(a, m):
    for i in range(1, m):
        if (a * i) % m == 1:
            return i
    return None

def decrypt(encrypted_text, a, b, m, alphabet):
    decrypted_text = ""
    a_inverse = mod_inverse(a, m)
    
    if a_inverse is not None:
        for char in encrypted_text:
            if char in alphabet:
                index = alphabet.index(char)
                decrypted_index = (a_inverse * (index - b)) % m
                decrypted_text += alphabet[decrypted_index]
            else:
                decrypted_text += char
        return decrypted_text
    else:
        return "No se puede calcular el inverso multiplicativo para la clave dada."

alphabet = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"

m = len(alphabet)

with open("cipher2.txt", "r") as file:
    encrypted_text = file.read()

for a in range(1, m):
    for b in range(1, m):
        decrypted_text = decrypt(encrypted_text, a, b, m, alphabet)
        print(f"Clave: a={a}, b={b} - Texto descifrado: {decrypted_text}")
```
